# Getting Candiadate Embeddings

```python
models = ["jinaai/jina-embeddings-v3",
           "Snowflake/snowflake-arctic-embed-l-v2.0", "BAAI/bge-m3",
           'sentence-transformers/all-MiniLM-L6-v2',
           'sentence-transformers/all-mpnet-base-v2',
           'Qwen/Qwen3-Embedding-0.6B', 'google/embeddinggemma-300m']
```

```python
import pandas as pd
df = pd.read_parquet("data/migration.parquet")
```

```python
from sentence_transformers import SentenceTransformer
for model in models:
  print(model)
  model_st = SentenceTransformer(model, trust_remote_code=True)
  embeddings = model_st.encode(df['translatedText'].to_list())
  del model_st
  df[model] = list(embeddings)
```

## Google Embeddings with different prompts

```python
model = 'google/embeddinggemma-300m'

model_st = SentenceTransformer(model, trust_remote_code=True)
```

```python
prompts = ['document', 'Clustering', 'Classification', 'STS', 'Retrieval', 'Summarization']
```

```python
for prompt_name in prompts:
  print(prompt_name)
  embeddings = model_st.encode(df['translatedText'].to_list(), prompt=model_st.prompts[prompt_name])
  df[prompt_name] = list(embeddings)
```

```python
df.to_parquet("data/migration_with_embeddings.parquet")
```
